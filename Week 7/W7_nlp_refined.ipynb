## 2. spaCy: Modern NLP Pipelines

[spaCy](https://spacy.io/) is a modern, high-performance library for NLP in Python. Designed for industrial-strength applications, it provides efficient and accurate text processing pipelines covering tokenisation, POS tagging, dependency parsing, and named entity recognition.

**Core Components:**
- **Doc**: The entire processed text, containing all annotations
- **Token**: Each word or punctuation mark, with linguistic metadata

| `token.text` | The original word |

See [spaCy 101](https://spacy.io/usage/spacy-101) for documentation.

## 3. Word Embeddings: Words as Vectors

In linguistics, the meaning of a word encompasses its definitions, connotations, and relationships with other words. For computers, capturing this information numerically is challenging.

The breakthrough came from **distributional semantics**: the idea that a word's meaning can be understood by the company it keeps.

> *"You shall know a word by the company it keeps."*  
> — J.R. Firth (1957)

**Word embeddings** are trained by algorithms that adjust word vectors based on their co-occurrence in large text corpora. Words appearing in similar contexts end up positioned closely in the vector space. This captures linguistic relationships in a multidimensional geometric space.

**Key insight:** Words are represented as dense vectors (typically 50-300 dimensions), where similar words have similar vectors.

**Must-see resource:** For an in-depth explanation of Word2Vec and GloVe algorithms, see [Lena Voita's NLP Course](https://lena-voita.github.io/nlp_course/word_embeddings.html).


### Why This Matters for Political Science

Word embeddings enable quantitative analysis of concepts that were previously difficult to measure:
- **Ideology detection**: Words used by different political groups cluster differently
- **Framing analysis**: How the same issue is described differently across sources
- **Historical change**: How word meanings evolve over time (e.g., "liberal")
- **Bias detection**: Uncovering implicit associations in text corpora

## 4. Gensim: Working with Word Embeddings

[Gensim](https://radimrehurek.com/gensim/) is a Python library for topic modelling and vector-based text representation. It provides:

**Key Methods We'll Use:**

| Method | What it does |

### Visualising Embeddings with t-SNE

Word embeddings exist in high-dimensional space (100 dimensions in our case). To visualise them, we use t-SNE (t-distributed Stochastic Neighbor Embedding) to reduce to 2D while preserving local structure.


````python
#@title Visualise Embeddings in 2D

# Get top 500 most frequent words
top_words = model.index_to_key[:500]
embeddings = np.asarray([model.get_vector(word) for word in top_words])

# Reduce to 2D with t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
reduced_embeddings = tsne.fit_transform(embeddings)

# Create interactive plot
fig = go.Figure(data=[go.Scatter(
    x=reduced_embeddings[:, 0],
    y=reduced_embeddings[:, 1],
    mode='markers+text',
    text=top_words,
    textposition="bottom center",
    marker=dict(size=5, opacity=0.7),
    textfont=dict(size=8)
)])

fig.update_layout(
    title='2D t-SNE Visualisation of Top 500 Words',
    xaxis_title='t-SNE Dimension 1',
    yaxis_title='t-SNE Dimension 2',
    height=600
)

fig.show()
````

## 5. Vector Arithmetic and Semantic Analogies

One of the most fascinating properties of word embeddings is that they support meaningful arithmetic. The classic example:

$$\vec{king} - \vec{man} + \vec{woman} \approx \vec{queen}$$

This works because the vector from "man" to "king" captures the concept of "royalty", and adding this to "woman" lands near "queen".


### Quick Refresher: Vectors and Dot Products

If vectors feel rusty, here are the essentials:

**Vector** = a list of numbers, like `[3, 4]` or `[0.5, -0.2, 0.8]`

You can think of a vector as an arrow pointing somewhere in space. A 2D vector like `[3, 4]` points to coordinates (3, 4) on a flat plane. Word embeddings use 100+ dimensions, but the principle is the same.

**Magnitude (length)** = how long the arrow is
- For `[3, 4]`: length = √(3² + 4²) = √25 = 5
- Think of it as the distance from the origin to the point

**Dot product** = multiply matching positions, then sum
- `[1, 2, 3] · [4, 5, 6] = (1×4) + (2×5) + (3×6) = 4 + 10 + 18 = 32`

**Key insight:** The dot product is large when vectors point in similar directions, small (or zero) when they point at right angles, and negative when they point in opposite directions.

That's all you need to follow along. The libraries handle the calculations for us.


### Cosine Similarity: Measuring Word Relatedness

**The intuition:** Think of word embeddings as arrows pointing in different directions in space. Two words are similar if their arrows point in roughly the same direction.

**What does "direction" mean?**

Imagine two arrows starting from the same point:

```
    Same direction      Right angles       Opposite directions
         ↗ ↗              ↗                    ↗
        /  /             /                    /
       /  /             →                    ↙
    (similar)        (unrelated)         (opposites)
```

Cosine similarity measures the angle between two arrows:
- **1** = same direction (identical meaning)
- **0** = perpendicular (unrelated)
- **-1** = opposite directions (opposite meaning)

**Why not just use distance?**

Regular distance cares about both direction AND length. But word vectors can have different lengths depending on how common a word is. Cosine similarity ignores length and focuses only on direction, which better captures meaning.

**The formula (for reference):**

$$\text{cosine similarity}(A, B) = \frac{A \cdot B}{\|A\| \times \|B\|}$$

This divides the dot product by both magnitudes, which "cancels out" the length and leaves only the directional component.

Don't worry about computing this by hand. Gensim does it for us with `model.similarity()`.


````python
# Vector arithmetic examples
print("Analogy: king - man + woman = ?")
result = model.most_similar(positive=["king", "woman"], negative=["man"], topn=3)
for word, score in result:
    print(f"  {word}: {score:.3f}")
````

````python
# More analogies
print("\nAnalogy: paris - france + germany = ?")
result = model.most_similar(positive=["paris", "germany"], negative=["france"], topn=3)
for word, score in result:
    print(f"  {word}: {score:.3f}")
````

## 6. Semantic Projection

**The intuition:** Imagine you want to measure how "powerful" different political roles are. You could create a ruler with "weak" on one end and "strong" on the other, then place each role somewhere on that ruler.

Semantic projection does exactly this with word vectors.

[Semantic projection](https://www.nature.com/articles/s41562-022-01316-8) (Grand et al. 2022) is a method for mapping words onto interpretable semantic scales.

**Step 1: Define a dimension using opposites**

To create a "power" scale, we use word pairs that represent opposite ends:
- weak ↔ strong
- powerless ↔ powerful  
- follower ↔ leader

**Step 2: Compute the direction**

For each pair, we subtract one vector from the other. This gives us an arrow pointing from one concept towards the other:

```
vector("strong") - vector("weak") = "power direction"
```

We average across multiple pairs to get a stable, reliable direction.

**Step 3: Project words onto the scale**

Now we can take any word (like "president" or "citizen") and ask: how far along the weak→strong direction does this word fall?

This uses the **dot product**, which measures how much one vector aligns with another direction. A high positive value means the word aligns with the "strong" end. A negative value means it's closer to "weak".

**Visual analogy:**

```
weak ←───────────────────────────────────→ strong
         │           │              │
      citizen      mayor        president
```

**Why is this useful for research?**

You can create dimensions for any concept you can define with opposites:
- Political ideology: liberal ↔ conservative
- Sentiment: negative ↔ positive
- Formality: casual ↔ formal
- Agency: passive ↔ active

Then project politicians, policies, or media sources onto these scales to quantify abstract concepts that would otherwise be difficult to measure.


````python
# Define semantic dimensions using antonym pairs

gender_pairs = [
    ("boy", "girl"),
    ("man", "woman"),
    ("father", "mother"),
    ("husband", "wife"),
    ("king", "queen"),
    ("brother", "sister")
]

age_pairs = [
    ("young", "old"),
    ("child", "adult"),
    ("boy", "man"),
    ("girl", "woman"),
    ("infant", "elderly")
]

power_pairs = [
    ("weak", "strong"),
    ("powerless", "powerful"),
    ("subordinate", "dominant"),
    ("follower", "leader")
]
````

````python
#@title Helper Functions for Semantic Projection

def compute_dimension_vector(pairs, model):
    """Compute a unit vector representing a semantic dimension from word pairs."""
    vectors = []
    for word1, word2 in pairs:
        if word1 in model and word2 in model:
            vectors.append(model[word1] - model[word2])
    
    if not vectors:
        return None
    
    # Average and normalise to unit vector
    dim_vector = np.mean(vectors, axis=0)
    dim_vector = dim_vector / np.linalg.norm(dim_vector)
    return dim_vector


def project_words(words, dim_vector, model):
    """Project words onto a semantic dimension."""
    projections = {}
    for word in words:
        if word in model:
            projections[word] = np.dot(model[word], dim_vector)
    return projections


def visualise_projection_1d(projections, title="Semantic Projection"):
    """Visualise words on a 1D scale."""
    words = list(projections.keys())
    values = list(projections.values())
    
    # Sort by projection value
    sorted_pairs = sorted(zip(words, values), key=lambda x: x[1])
    words, values = zip(*sorted_pairs)
    
    plt.figure(figsize=(10, 6))
    plt.barh(words, values)
    plt.xlabel("Projection Value")
    plt.title(title)
    plt.tight_layout()
    plt.show()
````

````python
# Project political terms onto the power dimension
power_vector = compute_dimension_vector(power_pairs, model)

political_terms = [
    "president", "citizen", "voter", "senator", "minister",
    "protester", "activist", "bureaucrat", "dictator", "diplomat",
    "rebel", "monarch", "peasant", "governor", "mayor"
]

projections = project_words(political_terms, power_vector, model)
visualise_projection_1d(projections, "Political Terms on Power Dimension (weak → strong)")
````

### 3D Semantic Projection

We can project words onto two dimensions simultaneously, creating a 2D or 3D visualisation.


````python
#@title 3D Visualisation Functions

def compute_residual_vector(pairs, x_vector, y_vector, model):
    """Compute a residual dimension orthogonal to x and y."""
    vectors = []
    for word1, word2 in pairs:
        if word1 in model and word2 in model:
            v = model[word1] - model[word2]
            # Remove x and y components
            v = v - np.dot(v, x_vector) * x_vector
            v = v - np.dot(v, y_vector) * y_vector
            vectors.append(v)
    
    if not vectors:
        return None
    
    residual = np.mean(vectors, axis=0)
    residual = residual / np.linalg.norm(residual)
    return residual


def visualise_3d_projection(words, x_pairs, y_pairs, x_label, y_label, model):
    """Create a 3D visualisation of words projected onto two semantic dimensions."""
    x_vector = compute_dimension_vector(x_pairs, model)
    y_vector = compute_dimension_vector(y_pairs, model)
    z_vector = compute_residual_vector(x_pairs + y_pairs, x_vector, y_vector, model)
    
    # Compute coordinates
    x_coords, y_coords, z_coords = [], [], []
    valid_words = []
    
    for word in words:
        if word in model:
            valid_words.append(word)
            x_coords.append(np.dot(model[word], x_vector))
            y_coords.append(np.dot(model[word], y_vector))
            z_coords.append(np.dot(model[word], z_vector))
    
    # Normalise to [-1, 1] range for visualisation
    def normalise(arr):
        arr = np.array(arr)
        return 2 * (arr - arr.min()) / (arr.max() - arr.min()) - 1
    
    x_coords = normalise(x_coords)
    y_coords = normalise(y_coords)
    z_coords = normalise(z_coords)
    
    # Create 3D plot
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    ax.scatter(x_coords, y_coords, z_coords, c='blue', s=50)
    
    for word, x, y, z in zip(valid_words, x_coords, y_coords, z_coords):
        ax.text(x, y, z, word, fontsize=9)
    
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_zlabel('Residual')
    ax.set_title(f'3D Semantic Projection: {x_label} vs {y_label}')
    
    plt.tight_layout()
    plt.show()
````

````python
# Visualise gendered terms on gender × age dimensions
gendered_words = [
    "boy", "girl", "man", "woman",
    "father", "mother", "brother", "sister",
    "husband", "wife", "grandfather", "grandmother",
    "king", "queen", "prince", "princess",
    "uncle", "aunt", "nephew", "niece"
]

visualise_3d_projection(
    gendered_words, 
    gender_pairs, 
    age_pairs, 
    "Gender (male → female)", 
    "Age (young → old)", 
    model
)
````
