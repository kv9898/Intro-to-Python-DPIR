# **1. Why Do We Need Classes?**

One of the biggest challenges in programming is managing complexity. Object-oriented programming addresses this by letting you bundle **data** (attributes) and **behaviour** (methods) into cohesive units that represent real-world or conceptual entities.

In Python, classes are optional. You can write perfectly fine code using just functions and modules. But classes shine when you want to:

- **Group related data and functionality** in one place
- **Reuse and extend** existing code through inheritance
- **Model real-world relationships** (a President *is a* Politician, a Car *has an* Engine)

# **2. Defining a Class: Instances and Attributes**

A **class** is a blueprint for creating objects (also called **instances**). Each class can have:

- **Instance attributes**: data that belongs to each individual object (different for every instance)
- **Class attributes**: data shared across all instances of the class


### The `__init__()` method

The `__init__()` method (pronounced "dunder init") runs automatically when you create a new instance. Its job is to set up the instance's initial attributes. The first parameter is always `self`, which refers to the instance being created.


````python
class Politician:
    """A simple model of a politician."""

    # Class attribute: shared by ALL instances
    instance_count = 0

    def __init__(self, party, age):
        # Instance attributes: unique to EACH instance
        self.party = party
        self.age = age
        self.name = "generic_name"
        Politician.instance_count += 1   # update the shared counter


# Create two instances
p1 = Politician("Labour", 42)
p2 = Politician("Conservative", 55)

# Access instance attributes
print(p1.party, p1.age)        # Labour 42
print(p2.party, p2.age)        # Conservative 55

# Access class attribute (same for both)
print(Politician.instance_count)  # 2
print(p1.instance_count)          # 2 (looked up via the class)
````

> **Tip:** While you *can* add attributes outside of `__init__()`, initialising everything inside it keeps your class readable and predictable.


### A note on naming conventions

- **Class names** use CamelCase: `Politician`, `DataLoader`, `ElectionResult`
- **Instance and variable names** use snake_case: `my_politician`, `vote_count`
- **Every class should have a docstring** immediately after the class definition
- Use one blank line between methods within a class, two blank lines between classes

See the [PEP 8 Style Guide](https://peps.python.org/pep-0008/) for the full conventions.

# **3. Methods and Special Methods**

Methods are functions that belong to a class and operate on its instances. There are three main types:

| Type | Decorator | First parameter | Purpose |
|------|-----------|----------------|----------|
| Instance method | (none) | `self` | Operate on a specific instance |
| Class method | `@classmethod` | `cls` | Operate on the class itself |
| Special method | (none, dunder names) | `self` | Integrate with Python's built-in operations |


### Instance methods

Instance methods always take `self` as their first parameter, giving them access to the instance's attributes:


````python
class Politician:
    """A politician who can make speeches and age."""

    def __init__(self, party, age):
        self.party = party
        self.age = age

    def make_speech(self):
        """Print a campaign speech."""
        print(f"Vote for {self.party}!")

    def have_birthday(self):
        """Increment the politician's age by 1."""
        self.age += 1


p = Politician("Labour", 42)
p.make_speech()       # Vote for Labour!
p.have_birthday()
print(p.age)          # 43
````

### Special (dunder) methods

Python provides predefined methods with double-underscore names that let your class integrate with built-in operations:

- `__str__()` controls what `print(obj)` and `str(obj)` display
- `__len__()` lets you call `len(obj)`
- `__repr__()` controls the "developer-facing" representation
- `__call__()` makes an instance callable like a function


````python
class Politician:
    """A politician with special methods."""

    def __init__(self, party, age):
        self.party = party
        self.age = age

    def __str__(self):
        return f"Politician(party={self.party}, age={self.age})"

    def __len__(self):
        return len(self.__dict__)  # number of instance attributes


p = Politician("Labour", 42)
print(p)        # Politician(party=Labour, age=42)
print(len(p))   # 2
````

### Class methods

Class methods use the `@classmethod` decorator and take `cls` (the class itself) as their first parameter. They're useful for operations that apply to the class as a whole, not a specific instance:


````python
class Politician:
    """A politician with a class method."""

    instance_count = 0

    def __init__(self, party, age):
        self.party = party
        self.age = age
        Politician.instance_count += 1

    @classmethod
    def show_count(cls):
        """Return the number of Politician instances created."""
        return cls.instance_count


p1 = Politician("Labour", 42)
p2 = Politician("Conservative", 55)

print(Politician.show_count())  # 2
````

# **4. Inheritance**

When one class **inherits** from another, it takes on the parent's attributes and methods. The child class can then add new functionality or **override** existing methods.

This models "is-a" relationships: a President **is a** Politician, so `President` should inherit from `Politician`.


````python
class Politician:
    """Base class for politicians."""

    def __init__(self, party, age):
        self.party = party
        self.age = age

    def make_speech(self):
        print(f"Vote for {self.party}!")

    def __str__(self):
        return f"Politician(party={self.party}, age={self.age})"


class President(Politician):
    """A president inherits from Politician and adds tenure tracking."""

    def __init__(self, party, age, country):
        super().__init__(party, age)    # call the parent's __init__
        self.country = country
        self.tenure = 0

    def make_policy(self):
        return f"President of {self.country} signed a new policy."


pres = President("Democrats", 81, "USA")

# Inherited from Politician
pres.make_speech()         # Vote for Democrats!
print(pres.party)          # Democrats

# New in President
print(pres.make_policy())  # President of USA signed a new policy.
print(pres.country)        # USA

# isinstance checks
print(isinstance(pres, President))    # True
print(isinstance(pres, Politician))   # True (a President IS a Politician)
````

### The `super()` function

The `super()` function calls a method from the parent class. It is most commonly used in `__init__()` to ensure the parent's setup logic runs before the child adds its own attributes.

### Overriding methods

To change a parent method's behaviour, define a method with the same name in the child class. Python will use the child's version:


````python
class President(Politician):
    """A president with an overridden __str__."""

    def __init__(self, party, age, country):
        super().__init__(party, age)
        self.country = country

    def __str__(self):
        # Override the parent's __str__
        return f"President of {self.country} ({self.party}, age {self.age})"

    def make_speech(self):
        # Override the parent's make_speech
        print(f"As President of {self.country}, I promise to lead!")


pres = President("Democrats", 81, "USA")
print(pres)          # President of USA (Democrats, age 81)
pres.make_speech()   # As President of USA, I promise to lead!
````

# **5. Composition**

Composition models "has-a" relationships: instead of inheriting, one class **contains instances of another** as attributes.

This is useful when objects work together but aren't variations of the same thing. A Parliament **has** Politicians (it is not *a type of* Politician). A Car **has** an Engine.

| Relationship | Pattern | Example |
|------|------|-------|
| "is-a" | Inheritance | President **is a** Politician |
| "has-a" | Composition | Parliament **has** Politicians |


````python
class Party:
    """A political party with a policy position."""

    def __init__(self, name, policy_preference):
        self.name = name
        self.policy_preference = policy_preference

    def __str__(self):
        return f"Party({self.name})"


class Politician:
    """A politician who belongs to a party (composition)."""

    def __init__(self, age, party):
        self.age = age
        self.party = party       # party is a Party instance
        self.vote = False

    def cast_vote(self, proposition):
        """Vote yes if the proposition matches the party's preference."""
        if proposition == self.party.policy_preference:
            self.vote = True

    def make_speech(self):
        print(f"Vote for {self.party.name}!")

    def __str__(self):
        return f"Politician(age={self.age}, party={self.party.name})"


# A Politician HAS A Party
labour = Party("Labour", "Increase NHS funding")
p1 = Politician(42, labour)

p1.make_speech()   # Vote for Labour!
print(p1)          # Politician(age=42, party=Labour)

# The politician can access their party's data
print(p1.party.policy_preference)  # Increase NHS funding
````

# **Summary**

| Concept | What it does | Example |
|---------|-------------|----------|
| **Class** | Blueprint for creating objects | `class Politician:` |
| **Instance** | A specific object created from a class | `p = Politician("Labour", 42)` |
| **Instance attribute** | Data unique to each instance | `self.party = party` |
| **Class attribute** | Data shared across all instances | `instance_count = 0` |
| **Instance method** | Function operating on an instance (takes `self`) | `def make_speech(self):` |
| **Class method** | Function operating on the class (takes `cls`) | `@classmethod def show_count(cls):` |
| **Special method** | Dunder methods for built-in operations | `__str__`, `__len__`, `__init__` |
| **Inheritance** | Child class takes on parent's attributes/methods | `class President(Politician):` |
| **`super()`** | Calls the parent class's method | `super().__init__(party, age)` |
| **Override** | Child redefines a parent's method | Defining `__str__` in both classes |
| **Composition** | Object contains other objects as attributes | `self.party = Party(...)` |

````python
#@title Exercise 4 Solutions

# 4a)
class Engine:
    """A car engine."""

    def __init__(self, fuel_type):
        self.fuel_type = fuel_type

    def start(self):
        print(f"Engine started ({self.fuel_type})")


# 4b)
class Car:
    """A car composed with an Engine."""

    def __init__(self, make, model, engine):
        self.make = make
        self.model = model
        self.engine = engine  # composition: Car HAS AN Engine

    # 4c)
    def turn_on_ignition(self):
        print(f"{self.make} {self.model}:")
        self.engine.start()


# 4d)
petrol_car = Car("Toyota", "Corolla", Engine("petrol"))
electric_car = Car("Tesla", "Model 3", Engine("electric"))

petrol_car.turn_on_ignition()    # Toyota Corolla: Engine started (petrol)
electric_car.turn_on_ignition()  # Tesla Model 3: Engine started (electric)
````

````python
#@title Capstone Solutions

class Party:
    """A political party."""

    def __init__(self, name, policy_preference):
        self.name = name
        self.policy_preference = policy_preference


class Politician:
    """A politician who belongs to a party."""

    def __init__(self, age, party):
        self.age = age
        self.party = party
        self.vote = False

    def cast_vote(self, proposition):
        if proposition == self.party.policy_preference:
            self.vote = True

    def make_speech(self):
        print(f"Vote for {self.party.name}!")


class President(Politician):
    """A president (inherits from Politician)."""

    def __init__(self, age, party):
        super().__init__(age, party)
        self.tenure = 0

    def __str__(self):
        return f"President({self.party.name})"


# 5a-d)
class Parliament:
    """A parliament that holds votes on propositions."""

    max_members = 7

    def __init__(self, proposition):
        self.proposition = proposition
        self.members = []
        self.votes = []

    def add_member(self, politician):
        self.members.append(politician)
        if isinstance(politician, President):
            self.president = politician

    def hold_vote(self):
        for member in self.members:
            member.cast_vote(self.proposition)
            self.votes.append(member.vote)
        return sum(self.votes)

    def pass_law(self):
        if sum(self.votes) > Parliament.max_members // 2:
            return f"{self.proposition} is now law!"
        else:
            return f"{self.proposition} was rejected."


# 5e) Test
labour = Party("Labour", "Increase NHS funding")
tory = Party("Conservative", "Cut taxes")
libdem = Party("Lib Dem", "Increase NHS funding")

p1 = Politician(42, labour)
p2 = Politician(55, tory)
p3 = Politician(38, labour)
p4 = Politician(61, libdem)
p5 = President(67, labour)

parl = Parliament("Increase NHS funding")
for member in [p1, p2, p3, p4, p5]:
    parl.add_member(member)

yes_votes = parl.hold_vote()
print(f"Yes votes: {yes_votes}/{len(parl.members)}")
print(parl.pass_law())
````
